# Check for stationarity

```python
from statsmodels.tsa.stattools import adfuller

def ad_test(dataset):
    dftest = adfuller(dataset, autolag = 'AIC')
    print('1. ADF: ',dftest[0])
    print('2. P-Value: ',dftest[1])
    print('3. Num of Lags: ',dftest[2])
    print('4. Num of observations used for ADF Regression and Critical Values Calculation: ',dftest[3])
    print('5. Critical Values :')
    for key, val in dftest[4].items():
        print('\t',key, ': ',val)
```

```python
ad_test(data_["pm25_µg_m3"])
# How to interpret these results?
# we have to focus on the p-value, the smaller the p-value is, the more stationary the time series is.
# we can see that the time series is completely stationary since we have previosuly normalized the dataset so that
# weather is not an issue.
```

# Implementing the model

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
rows = data.shape[0]
train=data.iloc[0:int(0.65*rows),:]
test=data.iloc[int(0.65*rows):,:]
```

```python
model = ARIMA(train,order=(5,0,4))
model = model.fit()
model.summary()
```

Make predictions

```python
start = len(train)
end = len(train) + len(test)-1
pred = model.predict(start=start,end=end,typ='levels')
pred.index = test.index
print(pred)
pred = pd.DataFrame(pred)
```

```python
pred.index = pd.to_datetime(pred.index)
test_vs_pred=pd.concat([test,pred],axis=1)
```

```python
test_vs_pred = test_vs_pred.rename(columns={0: 'prediction'})
```

```python
test_vs_pred.plot(figsize=(50,5))
```

### ARIMA has well-known problems with seasonal time series if the seasonal cycle is "too long". Monthly data, with a seasonal length of 12 months, is fine. Weekly data, with a season of length 52 (disregarding fractional week numbers) are already a problem for ARIMA.

```python
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose

series = data
period = 8760 
results = seasonal_decompose(series, model='additive', period=period)
trend_estimate    = results.trend
periodic_estimate = results.seasonal
residual = results.resid
 
plt.figure(figsize=(500,10))
plt.subplot(411)
plt.plot(series,label='Original time series', color='blue')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend_estimate,label='Trend of time series',color='blue')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(periodic_estimate,label='Seasonality of time series',color='blue')
plt.legend(loc='best')
plt.subplot(414)
```
